```
# ONLY RUN THIS CELL ONCE!

import re

for i in site_assessment.index:
  villages = []
  cells = []
  # split the strings by spaces so we can still refer to the punctuation
  words_and_punc = site_assessment.Cell[i].split(sep=" ")

  # split the strings into words and remove all punctuation
  words = re.findall(r"[\w']+", site_assessment.Cell[i])

  enum_words = enumerate(words)

  # First identify any instances of 'cell' and 'village'
  for num, word in enum_words:
    # get rid of any sector entries
    if word == "sector" or word == "Sector":
      words[num] = ""
      words[num-1] = ""
      
    # if word is village, capitalize the word before it and add it to villages unless the previous word is village, then delete
    elif word == "village" or word == "Village" or word == "vilage" or word == "villages":
      if words[num-1] == "village" or words[num-1] == "Village" or words[num-1] == "vilage" or words[num-1] == "villages":
        words[num] = ""
        

      else:
        village_name = words[num-1].capitalize()
        villages.append(village_name)
      
    # if word is cell, capitalize the word before it and add it to cells
    elif word == "cell" or word == "Cell":
      cell_name = words[num-1].capitalize()
      cells.append(cell_name)

  # SPLIT BY "IF VILLAGES == 0" AND "IF CELLS == 0" (those that did not have a number of cell/village labels equal to
  # the number of names I singled out and fixed manually)

  if len(villages) == 0 and len(cells) == 0:

  # if neither are true and the length of words is equal to one, append that word to cells
    if len(words) == 1:
      cells.append(words[0].capitalize())

    elif len(words) == 2:
      # identify parantheses so we know if there is both a cell and a village listed
      if '(' in words_and_punc or ')' in words_and_punc[-1]:
        #print("EXCEPTION: cell and village listed without labels(len 2): ", i, words_and_punc)
        cells.append(words[0].capitalize())
        villages.append(words[1].capitalize())
        
      else:
        cells.append(words[0].capitalize())
        cells.append(words[1].capitalize())

    elif len(words) == 3:
      # identify parantheses so we know if there is both a cell and a village listed
      if '(' in words_and_punc or ')' in words_and_punc[-1]:
        #print("EXCEPTION: cell and village listed without labels(len 3): ", i, words_and_punc)
        cells.append(words[0])
        villages.append(words[1].capitalize())
        villages.append(words[2].capitalize())
        
      else:
        cells.append(words[0].capitalize())
        cells.append(words[1].capitalize())
        cells.append(words[2].capitalize())

    elif len(words) == 0:
      # leave it empty
      pass

    else:
      # identify parantheses so we know if there is both a cell and a village listed
      if '(' in words_and_punc or ')' in words_and_punc[-1]:
        cells.append(words[0].capitalize())
        cells.append(words[1].capitalize())
        villages.append(words[2].capitalize())
        villages.append(words[3].capitalize())

      else:
        for y in range(len(words)):
          cells.append(words[y].capitalize())
        #print("EXCEPTION: no cells/villages, greater than 3 words: ", i, words_and_punc)

  # replace the cell rows with the cleaned output and add in the new village rows
  site_assessment['Cell'][i] = "-".join(cells)
  site_assessment['Village'][i] = "-".join(villages)

  
```
