```
# create cap function to capitalize the first letter after a dash('-')
s = 'a-bo'
def cap(string):
  for i, char in enumerate(string):
    if char == '-':
      string = string[:i+1] + string[i+1:].capitalize()

  return string

s = cap(s)
print(s)
```

```
# change all caps to just capitilizing the first letter
site_assessment['Sector'] = site_assessment['Sector'].str.capitalize()

# run cap function 
site_assessment['Sector'] = site_assessment['Sector'].apply(cap)
```
