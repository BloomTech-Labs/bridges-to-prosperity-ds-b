```
# now need to check the hyphenated values
# need to modify this function so that it checks for words inside of strings
def get_similar_hyphenated(word,possibilities):
  score = 0
  score2 = 0
  match = ""
  match2 = ""
  for x in possibilities:
    s = fuzz.partial_ratio(word, x)
    if s > score:
      score2 = score
      score = s
      match2 = match
      match = x

    elif score == 100 and s == 100:
      score2 = s
      match2 = x

  if score2 < 100:
    #print(f'{word} - {match}-{match2}: {score}-{score2}% similarity')
    pass

  return score, score2, match, match2, word
```
