```
from fuzzywuzzy import fuzz

# this function checks for words inside of strings
def get_similar_hyphenated(word,possibilities):
  score = 0
  score2 = 0
  match = ""
  match2 = ""
  for x in possibilities:
    s = fuzz.partial_ratio(word, x)
    if s > score:
      score2 = score
      score = s
      match2 = match
      match = x

    elif score == 100 and s == 100:
      score2 = s
      match2 = x

  if score2 < 100:
    #print(f'{word} - {match}-{match2}: {score}-{score2}% similarity')
    pass

  return score, score2, match, match2, word
```

```
village_count = 0
cell_count = 0
equal_count = 0

for i in df.index:
  x = df['Community Served 1'][i]
  
  # only run on cases where we have a cell id and a usable community served value
  if df.Cell_ID[i] is not "?" and len(x) > 2:
    sect_id = int(df['Sector_ID'][i])
    cell_id = int(df['Cell_ID'][i])

    poss_cells = names_codes[names_codes['Sect_ID'] == sect_id].Cell.unique()
    poss_vills = names_codes[names_codes['Cell_ID'] == cell_id].Village.unique()

    cellscore1, cellscore2, cellmatch1, cellmatch2, word = get_similar_hyphenated(x,poss_cells)
    villscore1, villscore2, villmatch1, villmatch2, word = get_similar_hyphenated(x,poss_vills)

    # set the conditionals so that if they're equal we set it to the cell ###### fix the Single Letter entries, ie "M"

    if villscore1 == cellscore1:
      
      if cellscore1 > 80:
        #match the cell id
        poss_rows = cell_key[cell_key['Sect_ID'] == sect_id]
        matching_row = poss_rows[poss_rows['Cell'] == cellmatch1]

        df.loc[i,'CS1_ID'] = matching_row.Cell_ID.item()

      else:
        print('== : no match')
        # leave as is     

    elif villscore1 > cellscore1:
      print('c',cellscore1, cellscore2, cellmatch1, cellmatch2, word)
      print('v',villscore1, villscore2, villmatch1, villmatch2, word)
      if villscore1 > 80:
        #match the village id
        poss_rows = names_codes[names_codes['Cell_ID'] == cell_id]
        matching_row = poss_rows[poss_rows['Village'] == villmatch1]

        df.loc[i,'CS1_ID'] = matching_row.Vill_ID.item()

      else:
        print('v > c : no match')
        # leave as is  

    else:
      if cellscore1 > 80:
        #match the cell id
        poss_rows = cell_key[cell_key['Sect_ID'] == sect_id]
        matching_row = poss_rows[poss_rows['Cell'] == cellmatch1]

        df.loc[i,'CS1_ID'] = matching_row.Cell_ID.item()

      else:
        print('c > v : no match')
        # leave as is

df.head()
```
