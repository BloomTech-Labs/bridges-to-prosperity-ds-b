```
# find a "similar spelling identifier" function
# fuzzywuzzy - https://www.datacamp.com/community/tutorials/fuzzy-string-python
from fuzzywuzzy import fuzz

# define function
# input = mismatched word, output = og word, score, closest word
def get_similar(word,possibilities):
  score = 0
  match = ""
  for x in possibilities:
    s = fuzz.ratio(word, x)
    if s > score:
      score = s
      match = x

  print(f'{word} - {match}: {score}% similarity')

# run on all unmatched entries
for x in y:
  get_similar(x,gov_sect)
```
