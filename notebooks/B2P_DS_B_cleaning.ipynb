```
df = pd.read_csv('B2P Dataset_2020.10 (2) (1).csv')
df.head()
```

```
df.drop('Bridge Opportunity: Comments', axis = 1, inplace=True)
```

```
df.dtypes
```

```
df.isnull().sum()
```

```
df.drop('Proposed Bridge Location (GPS) (Longitude)', axis=1, inplace=True)
```

```
df.drop('Name of nearest city', axis=1, inplace=True)
df.drop('Name of nearest paved or sealed road', axis=1, inplace=True)
```

```
df.dtypes
```

```
df.drop('Incident descriptions', axis=1, inplace=True)
df.drop('Notes on social information', axis=1, inplace=True)
df.drop('Bridge Opportunity: General Project Photos', axis=1, inplace=True)
df.drop('Bridge Opportunity: CaseSafeID', axis=1, inplace=True)
```

```
df.dtypes
```

```
df.drop('Estimated span (m)', axis=1, inplace=True)
```

```
df.head(5)
```

```
df.head(3)
```

```
df["Market access blocked by river"].loc[df['Market access blocked by river'].notnull()] = '1'
df['Education access blocked by river'].loc[df['Education access blocked by river'].notnull()] = '1'
df['Health access blocked by river'].loc[df['Health access blocked by river'].notnull()] = '1'
df['Other access blocked by river'].loc[df['Other access blocked by river'].notnull()] = '1'
```

```
df['Market access blocked by river'] = df['Market access blocked by river'].replace(np.nan, '0')
df['Education access blocked by river'] = df['Education access blocked by river'].replace(np.nan, '0')
df['Health access blocked by river'] = df['Health access blocked by river'].replace(np.nan, '0')
df['Other access blocked by river'] = df['Other access blocked by river'].replace(np.nan, '0')
```

```
df.head(3)
```

```
df['Market access blocked by river'] = df['Market access blocked by river'].astype(int)
df['Education access blocked by river'] = df['Education access blocked by river'].astype(int)
df['Health access blocked by river'] = df['Health access blocked by river'].astype(int)
df['Other access blocked by river'] = df['Other access blocked by river'].astype(int)
```

```
df.dtypes
```

```
df.drop('Bridge Type', axis=1, inplace=True)
df.drop('Proposed Bridge Location (GPS) (Latitude)', axis=1, inplace=True)
```
